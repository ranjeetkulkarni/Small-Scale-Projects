```python
# importing the libraries
import pandas as pd
import numpy as np
```

```python
letters = pd.read_csv(r"C:\Data science training\SVM\letterdata.csv")
letters.head()
```

```python
letters.describe()
#since there is not much difference in between the mean value of the variables ie they are not suffering with scaling issue henceforth no need to normalize the data
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
# splitting data into train and test
train,test = train_test_split(letters, test_size = 0.20)



```

```python
print(letters.shape)
print(train.shape)
print(test.shape)
```

```python
test.info()
```

```python
# segregating the input and output variable from train and test
train_X = train.iloc[:, 1:]
train_y = train.iloc[:, 0]
test_X  = test.iloc[:, 1:]
test_y  = test.iloc[:, 0]
```

```python
train_X
```

```python
# kernel = linear
model_linear = SVC(kernel = "linear")
model_linear.fit(train_X, train_y)
pred_test_linear = model_linear.predict(test_X)

res_linear=np.mean(pred_test_linear == test_y)
```

```python
pd.crosstab(pred_test_linear,test_y)
```

```python
# kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(train_X, train_y)
pred_test_rbf = model_rbf.predict(test_X)

res_rbf=np.mean(pred_test_rbf==test_y)
```

```python
help(SVC)
```

```python
# kernel = polynomial
model_poly = SVC(kernel = "poly")
model_poly.fit(train_X, train_y)
pred_test_poly= model_poly.predict(test_X)

res_poly=np.mean(pred_test_poly==test_y)
```

```python
# kernel = sigmoid
model_sigmoid = SVC(kernel = "sigmoid")
model_sigmoid.fit(train_X, train_y)
pred_test_sigmoid = model_sigmoid.predict(test_X)

res_sigmoid=np.mean(pred_test_sigmoid==test_y)
```

```python
pd.DataFrame({'Model':['Linear','RBF','Polynomial','Sigmoid'],'Accuracy':[res_linear,res_rbf,]})
```
