```python
import pandas as pd
import numpy as np

```

```python
wbcd = pd.read_csv("/content/wbcd.csv")

wbcd.head()
```

```python
# converting B to Benign and M to Malignant 
wbcd['diagnosis'] = np.where(wbcd['diagnosis'] == 'B', 'Benign ', wbcd['diagnosis'])
wbcd['diagnosis'] = np.where(wbcd['diagnosis'] == 'M', 'Malignant ', wbcd['diagnosis'])

```

```python
wbcd.head()
```

```python
wbcd = wbcd.iloc[:, 1:32] # Excluding id column

```

```python
wbcd.head()
```

```python
wbcd.describe()
```

```python
# Normalization function D:\upendra\Data Science Training\KNN\wbcd.csv
def norm_func(i):
    x = (i-i.min())	/ (i.max()-i.min())
    return (x)

```

```python
# Normalized data frame (considering the numerical part of data)
wbcd_n = norm_func(wbcd.iloc[:, 1:])
wbcd_n.describe()

```

```python
X = np.array(wbcd_n.iloc[:,:]) # Predictors 
Y = np.array(wbcd['diagnosis']) # Target 

```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python


knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, Y_train)

pred = knn.predict(X_test)
pred

```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, pred))
pd.crosstab(Y_test, pred, rownames = ['Actual'], colnames= ['Predictions']) 


```

```python
# error on train data
pred_train = knn.predict(X_train)
print(accuracy_score(Y_train, pred_train))
pd.crosstab(Y_train, pred_train, rownames=['Actual'], colnames = ['Predictions']) 


```

```python
# creating empty list variable 
acc = []

# running KNN algorithm for 3 to 50 nearest neighbours(odd numbers) and 
# storing the accuracy values

for i in range(3,50,2):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, Y_train)
    train_acc = np.mean(neigh.predict(X_train) == Y_train)
    test_acc = np.mean(neigh.predict(X_test) == Y_test)
    Difference=train_acc-test_acc
    acc.append([i,train_acc, test_acc,Difference])


```

```python
df=pd.DataFrame(acc)
```

```python
df.columns='K','train_acc','test_acc','difference'
```
