```python
import pandas as pd 
import numpy as np
```

```python

uni1 = pd.read_excel(r"C:\Data science training\PCA\University_Clustering.xlsx")
uni1

```

```python
uni = uni1.drop(["State"], axis = 1)

uni
```

```python
uni.corr()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 

# Considering only numerical data 
uni.data = uni.iloc[:, 1:]
```

```python
# Normalizing the numerical data 
uni_normal = scale(uni.data)
uni_normal

```

```python
help(PCA)
```

```python
pca = PCA()
pca_values = pca.fit_transform(uni_normal)

```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var

```

```python
pd.DataFrame(pca.components_).T # showing the weights
#pca.components_[0]
# Cumulative variance 

```

```python
var
```

```python
var1 = np.cumsum(np.round(var, decimals = 4) * 100)
var1



```

```python
# Variance plot for PCA components obtained 
plt.plot(var1, color = "red")
```

```python
# PCA scores
pca_values

```

```python
pca_data = pd.DataFrame(pca_values)
pca_data
```

```python
pca_data.columns = "comp0", "comp1", "comp2", "comp3", "comp4","comp5"
pca_data
```

```python
final = pd.concat([uni.Univ, pca_data.iloc[:, 0:3]], axis = 1)
final
```

```python

# Scatter diagram
import matplotlib.pylab as plt
plt.scatter(x = final.comp0, y = final.comp1)
```
